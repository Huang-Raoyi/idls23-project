```python
class CNN_Mail(torch.utils.data.Dataset):
    def __init__(self, root_dir, file_name):
        self.data = pd.read_csv(root_dir + file_name)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        return self.data.iloc[idx].article, self.data.iloc[idx].highlights
```

```python
from nltk.translate.bleu_score import corpus_bleu

# Example dataset of reference translations and machine-generated outputs
dataset = [
    {'reference': ['the', 'cat', 'is', 'on', 'the', 'mat'], 'output': ['the', 'cat', 'is', 'on', 'the', 'mat']},
    {'reference': ['there', 'is', 'a', 'cat', 'on', 'the', 'mat'], 'output': ['there', 'is', 'a', 'cat', 'on', 'the', 'mat']},
    {'reference': ['the', 'cat', 'is', 'on', 'the', 'mat'], 'output': ['the', 'cat', 'is', 'on', 'a', 'mat']},
    {'reference': ['there', 'is', 'a', 'cat', 'on', 'the', 'mat'], 'output': ['the', 'cat', 'is', 'on', 'the', 'carpet']}
]

# Extract references and outputs from dataset
references = [data_point['reference'] for data_point in dataset]
outputs = [data_point['output'] for data_point in dataset]

# Calculate corpus-level BLEU4 score for whole dataset
score = corpus_bleu(references, outputs, weights=(0.25, 0.25, 0.25, 0.25))

print(score)
```
